```python
import pandas as pd
import numpy as np

print("Libraries Loaded Successfully")
```

```python
crop = pd.read_csv(
    "../dataset/raw/crop_production.csv"
)

print(crop.shape)

crop.head()
```

```python
print(crop.isnull().sum())
```

```python
crop_clean = crop.dropna(
    subset=["Production", "Crop"]
)

print(crop_clean.shape)
```

```python
crop_clean.columns = (
    crop_clean.columns
    .str.strip()
    .str.lower()
)

print(crop_clean.columns.tolist())
```

```python
rainfall = pd.read_csv(
    "../dataset/raw/rainfall.csv",
    sep=";"
)

print(rainfall.shape)

rainfall.head()
```

```python
daily_cols = [
    '1st','2nd','3rd','4th','5th','6th','7th',
    '8th','9th','10th','11th','12th','13th',
    '14th','15th','16th','17th','18th','19th',
    '20th','21st','22nd','23rd','24th','25th',
    '26th','27th','28th','29th','30th','31st'
]

rainfall["monthly_rainfall"] = (
    rainfall[daily_cols]
    .sum(axis=1)
)

print(
    rainfall[
        ["state", "district", "month", "monthly_rainfall"]
    ].head()
)
```

```python
rainfall_clean = rainfall[
    [
        "state",
        "district",
        "month",
        "monthly_rainfall"
    ]
]
```

```python
rainfall_clean.columns = (
    rainfall_clean.columns
    .str.strip()
    .str.lower()
)

print(rainfall_clean.columns.tolist())
```

```python
rainfall_clean["district"] = (
    rainfall_clean["district"]
    .str.strip()
    .str.lower()
)

rainfall_clean["state"] = (
    rainfall_clean["state"]
    .str.strip()
    .str.lower()
)

print("Rainfall Standardization Complete")
```

```python
rainfall_final = (
    rainfall_clean
    .groupby(
        ["state", "district"]
    )["monthly_rainfall"]
    .mean()
    .reset_index()
)

print(rainfall_final.shape)

rainfall_final.head()
```

```python
population = pd.read_csv(
    "../dataset/raw/population.csv"
)

print(population.shape)

population.head()
```

```python
print(population.columns.tolist())
```

```python
population.columns = (
    population.columns
    .str.strip()
)

print(population.columns.tolist())
```

```python
population["Population"] = (
    population["Population"]
    .str.replace(",", "", regex=False)
    .astype(int)
)

population["Growth"] = (
    population["Growth"]
    .str.replace("%", "", regex=False)
    .astype(float)
)
```

```python
print(population.columns.tolist())
```

```python
population_clean = population[
    [
        "District",
        "State",
        "Population",
        "Growth",
        "Literacy"
    ]
]

population_clean.head()
```

```python
population_clean.columns = (
    population_clean.columns
    .str.strip()
    .str.lower()
)

print(population_clean.columns.tolist())
```

```python
population_clean["district"] = (
    population_clean["district"]
    .str.strip()
    .str.lower()
)

population_clean["state"] = (
    population_clean["state"]
    .str.strip()
    .str.lower()
)

print("Population Standardization Complete")
```

```python
crop_pop = pd.merge(
    crop_clean,
    population_clean,
    on=["state", "district"],
    how="inner"
)

print(crop_pop.shape)

crop_pop.head()
```

```python
final_dataset = pd.merge(
    crop_pop,
    rainfall_final,
    on=["state", "district"],
    how="inner"
)

print(final_dataset.shape)

final_dataset.head()
```

```python
def normalize_text(x):
    return (
        str(x)
        .lower()
        .strip()
        .replace("&", "and")
        .replace("-", " ")
        .replace(".", "")
    )

crop_clean["district"] = crop_clean["district"].apply(normalize_text)
crop_clean["state"] = crop_clean["state"].apply(normalize_text)

rainfall_final["district"] = rainfall_final["district"].apply(normalize_text)
rainfall_final["state"] = rainfall_final["state"].apply(normalize_text)

population_clean["district"] = population_clean["district"].apply(normalize_text)
population_clean["state"] = population_clean["state"].apply(normalize_text)

print("Normalization Complete")
```

```python
crop_districts = set(crop_clean["district"].unique())

rain_districts = set(rainfall_final["district"].unique())

pop_districts = set(population_clean["district"].unique())

print(
    "Crop districts:",
    len(crop_districts)
)

print(
    "Rain districts:",
    len(rain_districts)
)

print(
    "Population districts:",
    len(pop_districts)
)

print(
    "Common districts:",
    len(
        crop_districts &
        rain_districts &
        pop_districts
    )
)
```

```python
crop_pop = pd.merge(
    crop_clean,
    population_clean,
    on=["state", "district"],
    how="inner"
)

print(crop_pop.shape)
```

```python
final_dataset = pd.merge(
    crop_pop,
    rainfall_final,
    on=["state", "district"],
    how="inner"
)

print(final_dataset.shape)

final_dataset.head()
```

```python
final_dataset["demand"] = (
    final_dataset["population"] * 0.002
)
```

```python
final_dataset["risk_level"] = pd.cut(
    final_dataset["yield"],
    bins=[0, 1000, 3000, float("inf")],
    labels=["High", "Medium", "Low"]
)
```

```python
final_dataset.to_csv(
    "../dataset/processed/final_dataset.csv",
    index=False
)

print("Final Dataset Saved")
```
